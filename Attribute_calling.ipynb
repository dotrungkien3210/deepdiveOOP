Attribute: tất cả những gì mà có thể gọi được là một biến, là một hàm hay là một class

```python
class myClass:
    language = 'Python'
    def say_hello():
     print(f'Hello from {myClass.language} language')
```

```python
myClass.say_hello()
```

```python
getattr(myClass,'say_hello')
```

```python
my_obj = myClass()
```

```python
type(my_obj)
```

```python
isinstance(my_obj,myClass)
```
